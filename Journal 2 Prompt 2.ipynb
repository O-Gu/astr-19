```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0,1,num=100)
y_x= np.sin(x)
f_x= np.cos(x)
```

```python
f, axarr = plt.subplots(1,2)
axarr[0].plot(x,f_x)
axarr[0].set_title(r'$\cos(x)$')

axarr[1].plot(x,y_x)
axarr[1].set_title(r'$\sin(x)$')
f.subplots_adjust(wspace= 0.5)

plt.savefig("Prompt2.pdf",bbox_inches='tight', facecolor= 'lightsteelblue')
```
