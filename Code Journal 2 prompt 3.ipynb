```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x=np.random.uniform(0,1,1000)
num_bin=100

plt.hist( x,bins=num_bin,edgecolor="darkorange")
plt.ylabel('Number per bins')
plt.xlabel('Random numbers')

plt.savefig('Prompt3.pdf',bbox_inches='tight',facecolor='darkorange')
```
